```python
from http.client import responses
import urllib.request
import requests
import csv
#import pandas as pd
url='https://raw.githubusercontent.com/Yokhong/CIVL4210/refs/heads/main/H2_human_pose_in_3D_space/H1_assignment_data/testing_raw.csv'
urllib.request.urlretrieve(url,"/content/testing.csv")


csv_pathway = 'testing.csv'
txt_pathway = 'testing.txt'
start_row = 0
end_row = 16
start_col = 0
end_col = 2

with open(csv_pathway, "r", newline="") as csv_file:
    csv_reader = csv.reader(csv_file)
    data = []
    for i, row in enumerate(csv_reader):
        if i >= start_row and i <= end_row:
            row_data = row[start_col:end_col+1]
            data.append(row_data)
transposed_data = list(map(list, zip(*data)))

with open(txt_pathway, "w") as txt_file:
        for row_data in transposed_data:
            txt_file.write(" ".join(row_data) + "\n")
```
